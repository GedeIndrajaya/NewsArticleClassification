# Making the Dataframe

Randomizing the dataset so it doesn't mess up the testing on modeling.

# Data Cleaning

```python
#Cleaning the text
def clean_text(text):
    # Removing the symbols
    text = ' '.join(text)
    text = text.replace('Jakarta', '')
    text = text.replace('CNBC', '')
    text = text.replace('Indonesia', '')
    text = text.replace('@', '')
    text = text.replace('#', '')
    text = text.replace('https://', '')
    text = text.replace('.com', '')
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.replace('\n', '')
    text = text.replace('[', '')
    return text

df_news['text'] = df_news['text'].apply(clean_text)

#lowercasing the text
df_news['text'] = df_news['text'].str.lower()
```

```python
df_news.sample(3)
```

# Tokenization

```python
import nltk
nltk.download('punkt')
```

```python
def tokenize(text):
    sentences = nltk.sent_tokenize(text)
    tokens = []
    for sentence in sentences:
        tokens.extend(nltk.word_tokenize(sentence))

    return tokens

df_news['text'] = df_news['text'].apply(tokenize)

df_news
```

# Removing Stopwords & Stemming

```python
!pip install Sastrawi
```

Removing stopwords using Sastrawi

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

Creating the Stopword Remover and Stemmer

```python
#Stopword
stopword_factory = StopWordRemoverFactory()
stopword_remover = stopword_factory.create_stop_word_remover()

#Stemmer
stemmer_factory = StemmerFactory()
stemmer = stemmer_factory.create_stemmer()

```

```python
def remove_stopwords(text):
    text = stopword_remover.remove(' '.join(text))
    text = stemmer.stem(text)
    return text

df_news['text'] = df_news['text'].apply(remove_stopwords)
df_news
```

# Text Representation

Encoding the Label Column

```python
label_encoder = LabelEncoder()
df_news['Label'] = label_encoder.fit_transform(df_news['Label'])

df_news['Label'].value_counts()
```

Using Stratify so the train, test, split is shuffled and get a balanced data

```python
X_train, X_test, y_train, y_test = train_test_split(df_news['text'], df_news['Label'], test_size=0.3, random_state=42, stratify = df_news['Label'])
```

```python
X_train.shape
```

```python
y_train.shape
```

###Count Vectorizer

```python
vectorizer = CountVectorizer(min_df = 3, max_features=50, ngram_range=(1,1))
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)
```

```python
vectorizer.get_feature_names_out()
```

### Modelling for Count Vectorizer

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

modelSVC = SVC()
modelSVC.fit(X_train_counts, y_train)
y_pred = modelSVC.predict(X_test_counts)

print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

rf_classifier.fit(X_train_counts, y_train)
y_pred = rf_classifier.predict(X_test_counts)

accuracy1 = accuracy_score(y_test, y_pred)
conf_matrix1 = confusion_matrix(y_test, y_pred)
class_report1 = classification_report(y_test, y_pred)

print("Accuracy:", accuracy1)
print("Confusion Matrix:\n", conf_matrix1)
print("Classification Report:\n", class_report1)
```

### TFID

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the TF-IDF vectorizer on the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Transform the test data using the same vectorizer
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Now, X_train_tfidf and X_test_tfidf contain the TF-IDF representations of the text data

# You can access the feature names (words) corresponding to the columns in the TF-IDF matrix like this
feature_names = tfidf_vectorizer.get_feature_names_out()

# Optionally, convert X_train_tfidf and X_test_tfidf to dense arrays for easier inspection (not necessary for most use cases)
X_train_tfidf_dense = X_train_tfidf.toarray()
X_test_tfidf_dense = X_test_tfidf.toarray()
```

```python
svc_classifier = SVC()

# Train the classifier on the TF-IDF vectorized training data
svc_classifier.fit(X_train_tfidf, y_train)

# Make predictions on the TF-IDF vectorized test data
y_pred = svc_classifier.predict(X_test_tfidf)

# Evaluate the model
accuracy2 = accuracy_score(y_test, y_pred)
conf_matrix2 = confusion_matrix(y_test, y_pred)
class_report2 = classification_report(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy2)
print("Confusion Matrix:\n", conf_matrix2)
print("Classification Report:\n", class_report2)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the TF-IDF vectorized training data
rf_classifier.fit(X_train_tfidf, y_train)

# Make predictions on the TF-IDF vectorized test data
y_pred = rf_classifier.predict(X_test_tfidf)

# Evaluate the model
accuracy4 = accuracy_score(y_test, y_pred)
conf_matrix4 = confusion_matrix(y_test, y_pred)
class_report4 = classification_report(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy4)
print("Confusion Matrix:\n", conf_matrix4)
print("Classification Report:\n", class_report4)
```
